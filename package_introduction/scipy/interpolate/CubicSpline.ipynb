# 三次样条插值 [CubicSpline](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.CubicSpline.html#scipy-interpolate-cubicspline)

[CubicSpline]:https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.CubicSpline.html#scipy-interpolate-cubicspline
**本文全篇引用[CubicSpline]**

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
x = np.arange(10)  # 要求：一维、有限、严格递增
y = np.sin(x)
cs = CubicSpline(x, y)  # 返回PPoly对象
xs = np.arange(-0.5, 9.6, 0.1)
fig, ax = plt.subplots(figsize=(6.5, 4))
ax.plot(x, y, 'o', label='data')
ax.plot(xs, np.sin(xs), label='true')
ax.plot(xs, cs(xs), label="S")  # 
ax.plot(xs, cs(xs, 1), label="S'")  # 一阶导
ax.plot(xs, cs(xs, 2), label="S''")  # 二阶导
ax.plot(xs, cs(xs, 3), label="S'''")
ax.set_xlim(-0.5, 9.5)
ax.legend(loc='lower left', ncol=2)
plt.show()
```
