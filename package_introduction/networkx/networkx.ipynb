# networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
from rich import print
```

```python
G = nx.DiGraph()
# 增加节点
G.add_node('a')  # 添加一个节点1
G.add_nodes_from(['b', 'c', 'd', 'e'])  # 加点集合
# G.add_cycle(['f', 'g', 'h', 'j'])  # 加环
H = nx.path_graph(10)  # 返回由10个节点的无向图
G.add_nodes_from(H)  # 创建一个子图H加入G
# G.add_node(H)  # 直接将图作为节点
G.add_edge('b','c', name='b->c')

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='red')
edge_labels = nx.get_edge_attributes(G, 'name')
print(edge_labels)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
# nx.draw(G)
plt.show()
```

```python
from dataclasses import dataclass
@dataclass
class TestNode(object):
    name: str
    id: tuple

    def __hash__(self) -> int:
        return hash(self.id)

    def __eq__(self, __value: object) -> bool:
        return self.id == __value.id

    def __str__(self) -> str:
        return self.name
    
    def __repr__(self) -> str:
        return str(self)

G = nx.DiGraph()
# 增加节点
G.add_node(TestNode('test', (1,)))  # 添加一个节点1
G.add_node(TestNode('test2', (2,)))  # 添加一个节点1
G.add_nodes_from(['b', 'c', 'd', 'e'])  # 加点集合
# G.add_cycle(['f', 'g', 'h', 'j'])  # 加环
H = nx.path_graph(10)  # 返回由10个节点的无向图
G.add_nodes_from(H)  # 创建一个子图H加入G
# G.add_node(H)  # 直接将图作为节点
G.add_edge('b','c', name='b->c')
G.add_edge(TestNode('test2', (1,)), 'c', name='test->c')

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='red')
edge_labels = nx.get_edge_attributes(G, 'name')
print(edge_labels)
print(G.edges())
print(G.nodes())
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

node_labels = {TestNode('test', (1,)): 'test3'}
nx.draw_networkx_labels(G, pos, labels=node_labels)
# nx.draw(G)
plt.show()
```
